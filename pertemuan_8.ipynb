Karena sebelumnya kolom umur berisi data yang tidak valid seperti "21years", "NaN_string", atau nilai yang berbentuk teks, fungsi pd.to_numeric() akan mencoba mengubah semuanya menjadi angka. Jika ada nilai yang tidak bisa dikonversi, parameter errors='coerce' akan otomatis mengubah nilai tersebut menjadi NaN. Langkah ini penting agar kolom umur memiliki format yang konsisten dan bisa diproses dalam perhitungan statistik.

Setelah data berhasil dikonversi, perintah hisyam_abdulah['umur'].mean() digunakan untuk menghitung rata-rata umur mahasiswa. Fungsi .mean() akan mengabaikan nilai NaN, sehingga hasil rata-rata hanya dihitung berdasarkan nilai umur yang valid saja. Dengan cara ini, Anda dapat memperoleh gambaran umum mengenai umur mahasiswa dalam dataset setelah proses pembersihan data dilakukan.

```python
mapping_umur = {
    'twenty': 20,
    '21years': 21,
    'NaN_string': 25,
    -5: 25,
    120: 25,
    150: 25
}
```

Kode mapping_umur tersebut digunakan untuk membuat sebuah kamus (dictionary) yang berfungsi sebagai aturan pemetaan (mapping) untuk membersihkan data pada kolom umur. Dalam dataset, terdapat beberapa nilai umur yang tidak valid seperti "twenty", "21years", "NaN_string", serta nilai ekstrem seperti -5, 120, dan 150 yang tidak mungkin mewakili umur mahasiswa. Dengan membuat dictionary seperti ini, Anda dapat mengganti nilai-nilai tidak valid tersebut secara otomatis ke nilai yang lebih masuk akal, misalnya 25 sebagai nilai pengganti (default). Mapping ini sangat membantu dalam proses data cleaning karena membuat pembersihan data lebih efisien, konsisten, dan mudah dipelihara. Langkah berikutnya biasanya adalah menerapkan mapping ini menggunakan fungsi .replace(mapping_umur) pada kolom umur, sehingga kolom tersebut menjadi lebih rapi, valid, dan siap digunakan untuk analisis statistik atau visualisasi.

```python
hisyam_abdulah['umur'] = hisyam_abdulah['umur'].replace(mapping_umur)
```

Kode hisyam_abdulah['umur'] = hisyam_abdulah['umur'].replace(mapping_umur) digunakan untuk menerapkan aturan pemetaan yang telah dibuat dalam dictionary mapping_umur agar nilai-nilai umur yang tidak valid di dalam dataset dapat diperbaiki secara otomatis. Proses ini menggantikan nilai seperti "twenty", "21years", "NaN_string", serta angka tidak masuk akal seperti -5, 120, dan 150 dengan nilai pengganti yang lebih logis, misalnya 25. Dengan melakukan replace berbasis mapping, Anda dapat membersihkan data dengan cara yang lebih efisien dan konsisten, terutama jika dataset besar atau mengandung banyak variasi kesalahan data. Setelah proses ini, kolom umur menjadi lebih seragam dan akurat sehingga siap untuk digunakan dalam analisis lanjutan seperti perhitungan statistik, visualisasi distribusi umur, atau pemodelan data lainnya.
